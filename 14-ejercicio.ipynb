**La Danza de los Puntos Aleatorios**

**Enunciado:**

Imagina que tienes un grupo de partículas que se mueven aleatoriamente en un espacio bidimensional. Queremos visualizar cómo se dispersan estas partículas con el tiempo.

1.  **Generación de Puntos:**
    *   Utiliza NumPy para generar 100 puntos aleatorios (coordenadas x e y) dentro de un cuadrado de lado 2 centrado en el origen (es decir, las coordenadas x e y deben estar entre -1 y 1).
2.  **Simulación de Movimiento:**
    *   Crea una función que simule el movimiento de las partículas. En cada paso de tiempo, cada partícula se moverá una pequeña distancia aleatoria en ambas direcciones (x e y). La distancia máxima de movimiento en cada dirección será de 0.1 unidades.
3.  **Visualización con Pandas y Matplotlib:**
    *   Guarda las posiciones de las partículas en cada paso de tiempo (por ejemplo, 5 pasos) en un DataFrame de Pandas. Las columnas serán 'x' e 'y', y cada fila representará un paso de tiempo.
    *   Usa Matplotlib para crear una serie de gráficos (uno por cada paso de tiempo) que muestren la dispersión de las partículas. Cada gráfico debe mostrar la posición de todas las partículas en ese momento.
    *   Opcional: Agrega un título a cada gráfico que indique el paso de tiempo.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generación de Puntos
num_particulas = 100
puntos = np.random.uniform(-1, 1, size=(num_particulas, 2))

# 2. Simulación de Movimiento
def mover_particulas(puntos, max_movimiento=0.1):
    movimiento = np.random.uniform(-max_movimiento, max_movimiento, size=puntos.shape)
    return puntos + movimiento

# 3. Visualización
pasos_tiempo = 5
for paso in range(pasos_tiempo):
    puntos = mover_particulas(puntos)
    df = pd.DataFrame(puntos, columns=['x', 'y'])

    plt.figure()
    plt.scatter(df['x'], df['y'])
    plt.title(f'Paso de Tiempo: {paso + 1}')
    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```
