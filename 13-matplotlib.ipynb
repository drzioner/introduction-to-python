### Matplotlib

Matplotlib es una biblioteca de Python que permite crear gráficos y visualizaciones de datos de alta calidad. Es una herramienta muy útil para científicos de datos, ingenieros y analistas que necesitan representar sus datos de forma clara y efectiva.

```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# Crear el gráfico
plt.plot(x, y)

# Mostrar el gráfico
plt.show()
```

Algunos de los tipos de gráficos que se pueden crear con Matplotlib son:

- **Gráficos de líneas**: se utilizan para mostrar la evolución de una variable a lo largo del tiempo o para comparar dos o más variables.
- **Gráficos de barras**: se utilizan para comparar diferentes categorías o grupos.
- **Gráficos de dispersión**: se utilizan para mostrar la relación entre dos variables.
- **Histogramas**: se utilizan para mostrar la distribución de una variable.
- **Gráficos de tarta**: se utilizan para mostrar la proporción de diferentes categorías en un conjunto de datos.

Gráfico de líneas:

```python
import matplotlib.pyplot as plt

# Datos
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [1500, 1800, 1600, 2100, 2300]

# Crear el gráfico de líneas
plt.plot(meses, ventas)

# Añadir etiquetas y título
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas mensuales')

# Mostrar el gráfico
plt.show()
```

Gráfico de barras:

```python
import matplotlib.pyplot as plt

# Datos
ciudades = ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
habitantes = [3200000, 1600000, 800000, 700000]

# Crear el gráfico de barras
plt.bar(ciudades, habitantes)

# Añadir etiquetas y título
plt.xlabel('Ciudad')
plt.ylabel('Habitantes')
plt.title('Población de ciudades')

# Mostrar el gráfico
plt.show()
```

Gráfico de dispersión:

```python
import matplotlib.pyplot as plt

# Datos
edad = [25, 30, 35, 40, 45, 50]
ingresos = [40000, 50000, 60000, 70000, 80000, 90000]

# Crear el gráfico de dispersión
plt.scatter(edad, ingresos)

# Añadir etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.title('Relación entre edad e ingresos')

# Mostrar el gráfico
plt.show()
```

Histograma:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
datos = np.random.normal(170, 10, 250)  # Generar datos aleatorios

# Crear el histograma
plt.hist(datos)

# Añadir etiquetas y título
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.title('Distribución de alturas')

# Mostrar el gráfico
plt.show()
```

Gráfico de tarta:

```python
import matplotlib.pyplot as plt

# Datos
sistemas_operativos = ['Windows', 'macOS', 'Linux', 'Otros']
porcentajes = [70, 15, 10, 5]

# Crear el gráfico de tarta
plt.pie(porcentajes, labels=sistemas_operativos, autopct='%1.1f%%')

# Añadir título
plt.title('Cuota de mercado de sistemas operativos')

# Mostrar el gráfico
plt.show()
```
